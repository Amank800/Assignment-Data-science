```python
# Normalization function 
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
df=(df.iloc[:,1:])
df
```

```python

# Normalized data frame (considering the numerical part of data)
crime_norm = norm_func(df.iloc[:,1:])
crime_norm
```

```python
# create dendrogram
dendrogram = sch.dendrogram(sch.linkage(crime_norm, method='single'))
```

```python
# create dendrogram
dendrogram = sch.dendrogram(sch.linkage(crime_norm, method='average'))
```

```python
# create clusters
hc = AgglomerativeClustering(n_clusters=3, affinity = 'euclidean', linkage = 'single')
```

```python
clusters=pd.DataFrame(hc.fit_predict(crime_norm),columns=['Clusters'])
clusters['Clusters'].value_counts()
```

```python
# save clusters for chart
y_hc = hc.fit_predict(crime_norm)
Clusters=pd.DataFrame(y_hc,columns=['Clusters'])
Clusters.head()
```

```python
df['Clusters']=clusters
df.head()
```

```python
#Import the libraries
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Import .csv file and convert it to a DataFrame object
crime=pd.read_csv("crime_data.csv")

crime.head()
```

```python
crime.info()
```

```python
crime.drop(['Unnamed: 0'],axis=1,inplace=True)
```

```python
array=crime.values
array
```

```python
stscaler = StandardScaler().fit(array)
X = stscaler.transform(array)
X
```

```python
dbscan = DBSCAN(eps=0.9, min_samples=5)
dbscan.fit(X)
```

```python
#Noisy samples are given the label -1.
dbscan.labels_
```

```python
cl=pd.DataFrame(dbscan.labels_,columns=['cluster'])
cl
```

```python
pd.concat([crime,cl],axis=1)
```
