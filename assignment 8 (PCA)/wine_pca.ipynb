```python
import pandas as pd 
import numpy as np
import seaborn as sns
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale 
```

```python
df = pd.read_csv('wine.csv')
df.head()
```

```python
df.info()
```

```python
df1=df.iloc[:,1:]
df1
```

```python
df2 = df1.values
df2
```

```python
df3 = scale(df2)
df3
```

```python
pca=PCA(n_components=13)

data_pca=pca.fit_transform(df3)
data_pca
```

```python
pca.components_
```

```python
var=pca.explained_variance_ratio_
var
```

```python
var1=np.cumsum(np.round(var,4)*100)
var1
```

```python
final_df=pd.concat([df['Type'],pd.DataFrame(data_pca[:,0:3],columns=['PC1','PC2','PC3'])],axis=1)
final_df
```

```python
fig=plt.figure(figsize=(16,12))
sns.scatterplot(data=final_df)
```

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import normalize
```

```python
plt.figure(figsize=(10,8))
dendrogram=sch.dendrogram(sch.linkage(df3,'complete'))
```

```python
hc =AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='ward')
hc
```

```python
AgglomerativeClustering(n_clusters=3)
```

```python
y_hc = hc.fit_predict(df3)
Clusters=pd.DataFrame(y_hc,columns=['Clusters'])
Clusters
```

```python
wine=df.copy()
wine['h_clustersid']=hc.labels_
wine
```

```python
from sklearn.cluster import KMeans
```

```python
wcss = []
for i in range(1, 6):
    kmeans = KMeans(n_clusters=i,random_state=2)
    kmeans.fit(df3)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1, 6), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
clusters_new = KMeans(3, random_state=30).fit(df3)
clusters_new
```
